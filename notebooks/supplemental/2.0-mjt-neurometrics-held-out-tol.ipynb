morphs.data.load.ephys_data(shuffle=True) causes minor discrepancies in the fit error
This notebook quantifies it

```python
import morphs
```

```python
block_path = morphs.paths.blocks()[-1]
```

```python
block_path = '/mnt/cube/mthielk/analysis/B1096/kwik/Pen05_Rgt_AP2500_ML100__Site02_Z2500__B1096_cat_P05_S02_4'
```

```python
accuracies, cluster_accuracies = morphs.data.load.cluster_accuracies()
```

```python
psychometric_params = morphs.data.load.psychometric_params()
```

```python
good_clusters = morphs.data.accuracies.good_clusters(cluster_accuracies[block_path])
spikes = morphs.data.load.ephys_data(
        block_path, good_clusters=good_clusters, shuffle_endpoints=True)
```

```python
X, labels = morphs.spikes.create_neural_rep(spikes, max_id_len=5)
```

```python
representations = morphs.data.neurometric.logistic_dim_reduction(X, labels)
```

```python
behavior_subj = psychometric_params.keys()[0]
```

```python
label_df = morphs.data.neurometric.make_label_df(labels, behavior_subj, psychometric_params)
```

```python
behavior_df = morphs.data.neurometric.make_behavior_df(behavior_subj, psychometric_params)
```

```python
morphs.data.neurometric._calc_samples(representations, label_df, behavior_df, 0, tol=1e-4)
```

```python
morphs.data.neurometric._calc_samples(representations, label_df, behavior_df, 0, tol=1e-4)
```

```python
morphs.data.neurometric._calc_samples(representations, label_df, behavior_df, 0, tol=1e-16)
```

```python
morphs.data.neurometric.load_neurometric_null_block(block_path, 4, cluster_accuracies, psychometric_params)
```

```python
morphs.data.neurometric.generate_neurometric_null_block(block_path, 4, cluster_accuracies, psychometric_params)
morphs.data.neurometric.load_neurometric_null_block(block_path, 4, cluster_accuracies, psychometric_params)
```

```python
from joblib import Parallel, delayed
```

```python
all_samples = Parallel(n_jobs=morphs.parallel.N_JOBS)(delayed(morphs.data.neurometric._calc_samples)(representations, label_df, behavior_df,
                                                                     idx) for idx in range(4))
```
