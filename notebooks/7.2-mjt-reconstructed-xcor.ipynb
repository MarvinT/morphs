```python
from __future__ import division
import morphs
from morphs.data import xcor, derivative

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
import itertools
```

```python
def reconstruct_xcor(popt):
    sampled_points = list(itertools.combinations_with_replacement(np.arange(1, 129), 2))
    x, y = zip(*sampled_points) # because py2 doesn't do unpacking in a list :(
    xyz = np.array(
        [x, y, derivative.integrate_pairs(popt, sampled_points)]
    ).T
    xyz = xcor.xyz_vf2sf(xyz)
    morph_pos_list, grid = xcor.xyz_to_sp(xyz)
    return grid
```

```python
def plt_recon_xcor(grid, morph_dim="", xlabel="Morph Position", ylabel="Morph Position"):
    ax = plt.gca()
    ax.imshow(-grid, extent=(1, 128, 1, 128))
    morphs.plot.format_morph_dim_ax_label(ax, morph_dim=morph_dim)
    morphs.plot.format_morph_dim_ax_label(ax, morph_dim=morph_dim[::-1], x_axis=False)
    ax.set_ylabel(ylabel)
    ax.set_xlabel(xlabel)
    return ax
```

```python
dd = morphs.load.derivative_dict()
ddf = morphs.load.derivative_df()
```

```python
morph_dim = "ac"
for selected_block_path in dd:
    if (
        "Pen01_Lft_AP2500_ML500__Site06_Z3000__B1183_cat_P01_S06_5"
        in selected_block_path
    ):
        break
popt = dd[selected_block_path][morph_dim][4]
```

```python
grid = reconstruct_xcor(popt)
```

```python
ax = plt_recon_xcor(grid, morph_dim=morph_dim)
morphs.plot.savefig(plt.gcf(), "xcor_recon_block_ac")
```

```python
morph_dim = "ae"
count = 0
grid = np.zeros((128, 128))
for block_path in dd:
    popt = dd[block_path][morph_dim][4]
    grid += reconstruct_xcor(popt)
    count += 1
ax = plt_recon_xcor(grid / count, morph_dim=morph_dim)
morphs.plot.savefig(plt.gcf(), "xcor_recon_mean_ae")
```

```python
morph_dim = "ah"
count = 0
grid = np.zeros((128, 128))
for block_path in dd:
    popt = dd[block_path][morph_dim][4]
    grid += reconstruct_xcor(popt)
    count += 1
ax = plt_recon_xcor(grid / count, morph_dim=morph_dim, ylabel="")
morphs.plot.savefig(plt.gcf(), "xcor_recon_mean_ah")
```
